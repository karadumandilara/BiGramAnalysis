### Clean Function for clean the tweets from hashtags, RTs, links, and other handle patterns

```python
url_pattern = re.compile("((http|https|https://)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#])*")
handle_pattern = re.compile("(@[A-Za-z0-9Ä±Ã§Ã¶ÅŸÃ¼ÄŸÄ°Ã‡Ã–ÅžÄžÃœ_.' ']+.-',':.)")


def clean_text(text):
    text = str(text)
    #tokens = tknzr.tokenize(text) # metni token lara ayÄ±rÄ±p listeye Ã§evirdi
    tokens = text.split()
    
    # her bir kelimeyi kontrolden geÃ§irerek 
    clean_tokens = []
    for t in tokens:
        t = str(t)
        # remove stopwords
        if t in stop_words:
            continue

        #remove links
        t = url_pattern.sub(r'', t)

        # remove twitter handles?
        t = handle_pattern.sub(r'', t)
        
        # remove # and RT
        t = t.replace('#', '').replace('RT', '')

        t = t.strip() # saÄŸdan soldan boÅŸluklarÄ± alÄ±r.

        if t != '' or "\'" or '"\w+' or "," or "''" or " " or "'" or "." or ":":
            clean_tokens.append(t)

    # make text from tokens
    clean_text = " ".join(clean_tokens)
    return clean_text

tweets = clean_text(tweets)
#tweets = [tweets]
tweets = tweets.lower()
import string # remove punc.
tweets = tweets.translate(str.maketrans('', '', string.punctuation))
print(tweets)
tweets = re.sub(r'\w*\d\w*', '', tweets).strip() # remove meaningless words with numbers
```

```python
###### use it
people_use = re.compile("(\S+\s*people)")
result = re.findall(people_use, tweets)
result_dict2 = {}

for i in result:
    if i in result_dict2.keys():
        result_dict2[i] = result_dict2[i] + 1
    else:
        result_dict2[i] = 1
#sort the result_dict from big to small.      
for i in sorted (result_dict2) : 
    print ((i, result_dict2[i]), end =" ") 
result2 = pd.DataFrame.from_dict(result_dict2, orient='index')
result2 = result2.reset_index()
result2 = result2.sort_values(by=0, ascending = False)
result2 = result2.drop(1)
import seaborn as sns
sns.set(rc={'figure.figsize':(15,5)})
h=sns.barplot(x="index", y=0, data=result2.iloc[:30])
h.set_xticklabels(h.get_xticklabels(), rotation=90)
#result2 = result2.iloc[1].drop
result2
```
